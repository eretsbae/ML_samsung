# Decision Tree Regression

의사 결정 나무(decision tree)는 여러 가지 규칙을 순차적으로 적용하면서 독립 변수 공간을 분할하는 분류 모형이다.
분류(classification)와 회귀 분석(regression)에 모두 사용될 수 있기 때문에 CART(Classification And Regression Tree)라고도 한다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

다음과 같은 샘플데이터를 생성하여 Decision Tree Regression을 적용해 보도록 한다.

```python
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
```

```python
plt.plot(X, y, 'o')
plt.show()
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

regtree1 = DecisionTreeRegressor(max_depth=3)
regtree1.fit(X, y)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_hat = regtree1.predict(X_test)
```

```python
regtree1.score(X, y)
```

```python
# Plot the results
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_test, y_hat, color="cornflowerblue", linewidth=2, label="predict")
plt.xlabel("x")
plt.ylabel("Y value")
plt.title("Decision Tree Regression - max_depth=3")
plt.legend()
plt.show()
```

트리의 최대 깊이 ( max_depth 매개 변수에 의해 제어 됨 )가 너무 높게 설정되면 의사 결정 트리가 overfitting 되게 된다.

```python
regtree2 = DecisionTreeRegressor(max_depth=5)
regtree2.fit(X, y)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_hat = regtree2.predict(X_test)

plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_test, y_hat, color="cornflowerblue", linewidth=2, label="predict")
plt.xlabel("x")
plt.ylabel("Y value")
plt.title("Decision Tree Regression - max_depth=5")
plt.legend()
plt.show()
```
